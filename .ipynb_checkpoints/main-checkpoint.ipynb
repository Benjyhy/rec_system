```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn.cluster import KMeans
from sklearn.preprocessing import LabelEncoder
```

```python
# Dataset exploration
cols = ['ticket_id', 'mois', 'prix', 'famille', 'univers', 'maille', 'libelle', 'cli_id']
global_df = pd.read_csv("KaDo.part", names=cols, low_memory=False)
global_df = global_df[1:]
explore_df = global_df.drop(['ticket_id', 'cli_id'], axis=1)

def explore(df, verbose=False):
    for col in df.columns:
        plt.rcdefaults()
        fig, ax = plt.subplots()
        
        y_vals = df[col].value_counts().index.tolist()
        y_pos = np.arange(len(y_vals))
        x_vals = df[col].value_counts()
        
        ax.barh(y_pos, x_vals, align='center')
        ax.set_yticks(y_pos, labels=y_vals)
        ax.set_xlabel('Counts')
        ax.set_xlabel('Values')
        ax.set_title(col)
        
        plt.show()
        
        if verbose:
            print(f"""
                Number of {col} : {len(df[col].value_counts())}
                Possible values for {col} : {df[col].value_counts().index.tolist()}
            """)

explore(explore_df)

print(f"""
Number of items : {len(global_df)}
Number of clients : {len(global_df['cli_id'].value_counts())}
Number of tickets : {len(global_df['ticket_id'].value_counts())}
""")
```

```python
#K-means clustering with family and price
prix_famille_df = global_df[['famille', 'prix']]
print(prix_famille_df.head())

family_le = LabelEncoder()

family_le.fit(prix_famille_df['famille'].value_counts().index.tolist())
family_values = family_le.transform(prix_famille_df['famille'].values)
family_values = np.array(family_values)
price_values = np.array(prix_famille_df['prix'].values).astype(float)
X_prix_famille_km = np.column_stack((family_values,price_values))
kmeans = KMeans(n_clusters=10, random_state=0).fit(X_prix_famille_km.tolist())

plt.scatter(X_prix_famille_km[:, 0], X_prix_famille_km[:, 1], c=kmeans.labels_, s=50, cmap='viridis')
plt.scatter(kmeans.cluster_centers_[:, 0], kmeans.cluster_centers_[:, 1], c='black', s=200, alpha=0.5);
plt.show()

print(family_le)
```

```python
#K-means clustering with universe and price
prix_univers_df = global_df[['univers', 'prix']]
print(prix_univers_df.head())

univers_le = LabelEncoder()
print(len(prix_univers_df['univers'].value_counts().index.tolist()))
univers_le.fit(prix_univers_df['univers'].values)
univers_values = univers_le.transform(prix_univers_df['univers'].values)
univers_values = np.array(univers_values)
price_values = np.array(prix_univers_df['prix'].values).astype(float)
X_prix_univers_km = np.column_stack((univers_values,price_values))
kmeans_2 = KMeans(n_clusters=9, random_state=0).fit(X_prix_univers_km.tolist())

plt.scatter(X_prix_univers_km[:, 0], X_prix_univers_km[:, 1], c=kmeans_2.labels_, s=50, cmap='viridis')
plt.scatter(kmeans_2.cluster_centers_[:, 0], kmeans_2.cluster_centers_[:, 1], c='black', s=200, alpha=0.5);
plt.show()
```
